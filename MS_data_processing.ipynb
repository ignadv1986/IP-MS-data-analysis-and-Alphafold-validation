```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("MS dataset.xlsx")
df.head()
```

```python
c = 6
epsilon = 1e-6
df['WT/D1 fdr_threshold'] = c / (df['WT/D1'].abs() + epsilon)
df['WT/SANT fdr_threshold'] = c / (df['WT/SANT'].abs() + epsilon)
df['D1/SANT fdr_threshold'] = c / (df['D1/SANT'].abs() + epsilon)
df.head(10)
```

```python
def classify(row):
    if abs(row['WT/D1']) < 0.5:
        return 'Not Significant'
    elif row['WT/D1 p-value'] > row['WT/D1 fdr_threshold']:
        if row['WT/D1'] > 0:
            return 'Significant Right'
        else:
            return 'Significant Left'
    else:
        return 'Not Significant'

df['category WT/D1'] = df.apply(classify, axis=1)
df.head()
```

```python
def classify(row):
    if abs(row['WT/SANT']) < 0.5:
        return 'Not Significant'
    elif row['WT/SANT p-value'] > row['WT/SANT fdr_threshold']:
        if row['WT/SANT'] > 0:
            return 'Significant Right'
        else:
            return 'Significant Left'
    else:
        return 'Not Significant'

df['category WT/SANT'] = df.apply(classify, axis=1)
df.head()
```

```python
def classify(row):
    if abs(row['D1/SANT']) < 0.5:
        return 'Not Significant'
    elif row['D1/SANT p-value'] > row['D1/SANT fdr_threshold']:
        if row['D1/SANT'] > 0:
            return 'Significant Right'
        else:
            return 'Significant Left'
    else:
        return 'Not Significant'

df['category D1/SANT'] = df.apply(classify, axis=1)
df.head()
```

```python
WT_D1 = df[["Protein", "WT/D1","WT/D1 p-value", "WT/D1 fdr_threshold", "category WT/D1"]]
WT_D1 = WT_D1.sort_values(by = "category WT/D1")
WT_D1.head()
```

```python
WT_SANT = df[["Protein", "WT/SANT","WT/SANT p-value", "WT/SANT fdr_threshold", "category WT/SANT"]]
WT_SANT = WT_SANT.sort_values(by = "category WT/SANT")
WT_SANT.head()
```

```python
D1_SANT = df[["Protein", "D1/SANT","D1/SANT p-value", "D1/SANT fdr_threshold", "category D1/SANT"]]
D1_SANT = D1_SANT.sort_values(by = "category D1/SANT")
D1_SANT.head()
```

```python
with pd.ExcelWriter('volcano_all_comparisons.xlsx') as writer:
    WT_D1.to_excel(writer, sheet_name='WT_D1', index=False)
    WT_SANT.to_excel(writer, sheet_name='WT_SANT', index=False)
    D1_SANT.to_excel(writer, sheet_name='D1_SANT', index=False)
from google.colab import files
files.download('volcano_all_comparisons.xlsx')
```
